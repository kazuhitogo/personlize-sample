```python
import pandas as pd
import datetime
```

```python
interaction_df = pd.read_csv('./menu2020.csv')
interaction_df
```

```python
# interaction_out_df = interaction_df.rename(columns={'料理名': 'ITEM_ID'})
interaction_df['USER_ID'] = interaction_df['料理名'].map(lambda x: str(1))
interaction_df['TIMESTAMP_UNIX'] = interaction_df['TIMESTAMP'].map(lambda x: int(datetime.datetime.strptime(x, '%Y/%m/%d %H:%M:%S').timestamp()))
```

```python
interaction_df
```

```python
item_df = pd.read_csv('./menu_master.csv')
item_df
```

```python
item_df['ITEM_ID'] = item_df.index.values+1
```

```python
interaction_out_df = pd.merge(interaction_df,item_df,on='料理名')[['USER_ID','ITEM_ID','TIMESTAMP_UNIX']].rename(columns={'TIMESTAMP_UNIX': 'TIMESTAMP'})
interaction_out_df
```

```python
def make_genre(X):
    y = []
    for x in X[1:7]:
        if x != " ":
            y.append(x)
    return "|".join(y)
```

```python
item_df['GENRE'] = item_df.apply(make_genre,axis=1)
item_out_df = item_df.reindex(columns=['ITEM_ID', 'GENRE'])
```
